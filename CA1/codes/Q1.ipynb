```python
import re

def custom_tokenizer(text):
  pattern = r'\b\w+\b'
  tokens = re.findall(pattern, text)
  return tokens
```

```python
sentence = 'Just received my M.Sc. diploma today, on 2024/02/10! Excited to embark on this new journey of knowledge and discovery. #MScGraduate #EducationMatters.'
```

```python
custom_tokenizer(sentence)
```

```python
def my_custom_tokenizer(text):
  pattern = r'[^\w\s]|\b[\w\S]+\b'
  tokens = re.findall(pattern, text)
  return tokens
```

```python
my_custom_tokenizer(sentence)
```
